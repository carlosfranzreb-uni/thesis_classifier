Plot the training loss per epoch. Training this model took about 11 hours.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
log_file = '../logs/training_1643982969.log'
test_losses, tr_losses = [], []
with open(log_file) as f:
  for line in f.readlines():
    if 'test' in line:
      test_losses.append(float(line.split(': ')[-1]))
    elif 'loss of epoch' in line:
      tr_losses.append(float(line.split(': ')[-1]))
test_losses, tr_losses
```

```python
plt.scatter(range(1, len(test_losses)+1), test_losses)
plt.plot(range(1, len(test_losses)+1), test_losses)
plt.xlabel('Epoch')
plt.ylabel('Avg. testing loss')
plt.xticks(range(1, len(test_losses)+1))
plt.show()
```

```python
plt.scatter(range(1, len(tr_losses)+1), tr_losses)
plt.plot(range(1, len(tr_losses)+1), tr_losses)
plt.xlabel('Epoch')
plt.ylabel('Avg. training loss')
plt.xticks(range(1, len(tr_losses)+1))
plt.show()
```
